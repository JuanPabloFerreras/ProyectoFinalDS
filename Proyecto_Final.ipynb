```python
from cycler import mul
EDV2 = EDV [[ 'Country','Region','Year', 'GDP_per_capita','Economy_status_Developing','Population_mln','Life_expectancy']]
```

```python
column_names_mapping = {'Country': 'País',
                        'Region': 'Región',
                        'Year': 'Año',
                        'GDP_per_capita': 'Ingreso per capita',
                        'Economy_status_Developing  ': 'Estado Desarrollo Económico',
                        'Population_mln': 'Población',
                        'Life_expectancy':'Esperanza de Vida (EV)'}

EDV3 = EDV2.rename(columns=column_names_mapping)

EDV4 = EDV3.sort_values(by=['Año', 'Región', 'País'])

EDV4
```

```python
import pandas as pd

# Assuming you have a DataFrame named "table" with columns to group

# Group rows by a single column
grouped_EDV4 = EDV4.groupby( ['Año', 'Región'])

# Perform aggregation on the grouped data
EDV5 = grouped_EDV4.mean()

EDV5
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Región', y='Ingreso per capita', data=EDV5)
plt.title('Esperanza de vida por Región')
plt.xlabel('Region')
plt.ylabel('Tasa de EV')
plt.show()

```

Al realizar un análisis inicial del conjunto de datos original, se observa que los datos corresponden a variables medidas de la misma forma provienes de todas las regiones mundiales. En virtud de esto, no será necesario agregar información adicional al conjunto de datos para estandarizar.
